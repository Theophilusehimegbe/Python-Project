```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings 
warnings.filterwarnings('ignore')
```

```python
pd.read_csv('movies.csv')
```

```python
df = pd.read_csv('movies.csv')
```

```python
df.drop_duplicates(inplace=True)
```

```python
print("\nMissing Values:\n", df.isnull().sum())
```

```python
columns_to_drop = ['homepage', 'tagline', 'keywords']
df.drop(columns=[col for col in columns_to_drop if col in df.columns], inplace=True)
```

```python
irrelevant_columns = ['id', 'release_date', 'overview', 'cast', 'budget', 'revenue', 'poster_path']
df.drop(columns=[col for col in irrelevant_columns if col in df.columns], inplace=True)
```

```python
df.dropna(subset=['budget_adj', 'revenue_adj', 'runtime'], inplace=True)
```

```python
df
```

```python
df['popularity'] = pd.to_numeric(df['popularity'], errors='coerce')
```

```python
df = df.dropna(subset=['popularity'])
```

```python
df['popularity'] = df['popularity'].round().astype(int)
```

```python
df['release_year'] = df['release_year'].astype(int)
```

```python
df['vote_count'] = df['vote_count'].round().astype(int)
```

```python
df = df.dropna(subset=['imdb_id'])
```

```python
df['production_companies'] = df['production_companies'].fillna('Unknown')
```

```python
df['genres'] = df['genres'].replace('', 'Unknown')
df['genres'] = df['genres'].fillna('Unknown')
```

```python
df['director'] = df['director'].replace('', 'Unknown')
df['director'] = df['director'].fillna('Unknown')
```

```python
df.rename(columns={'\timdb_id': 'imdb_id'}, inplace=True)
```

```python
print(df.columns.tolist())
```

```python
df['runtime'] = df['runtime'].astype(int)
```

```python
df['revenue_adj'] = df['revenue_adj'].round().astype(int)
```

```python
df['budget_adj'] = df['budget_adj'].round().astype(int)
```

```python
df['genres'] = df['genres'].str.split('|').str[0]
```

```python
df['production_companies'] = df['production_companies'].str.split('|').str[0]
```

```python
df = df.dropna(subset=['revenue_adj', 'budget_adj'])
```

```python
df
```

```python
print((df['revenue_adj'] == 0).sum())  
```

```python
print(df['revenue_adj'].dtype)
```

```python
df['revenue_adj'] = pd.to_numeric(df['revenue_adj'], errors='coerce')
```

```python
df = df[(df['revenue_adj'].notna()) & (df['revenue_adj'] != 0)]
```

```python
df['budget_adj'] = pd.to_numeric(df['budget_adj'], errors='coerce')
df = df[(df['budget_adj'].notna()) & (df['budget_adj'] != 0)]
```

```python
df.to_csv('movies_cleaned.csv', index=False)
```
